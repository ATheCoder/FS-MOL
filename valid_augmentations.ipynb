```python
import selfies
import random
import re
from rdkit import Chem

def visualize_smiles(s):
    return Chem.MolFromSmiles(s)

example_molecule = 'CCC(COC(=O)[C@@H](NP(=O)(Oc1ccccc1)OC[C@H]1O[C@@]([C@@H]([C@@H]1O)O)(C#N)c1ccc2n1ncnc2N)C)CC'

selfie = selfies.encoder(example_molecule)

visualize_smiles(example_molecule)
```

```python
def augmentSmiles(sm, percent = 0.2):
    sel = selfies.encoder(sm)
    
    characters = re.findall('\[([^\]]+)\]', sel)
    
    drop_nodes = int(len(characters) * percent)
    
    for _ in range(drop_nodes):
        characters.pop(random.randrange(len(characters)))
    
    return selfies.decoder("".join([f'[{char}]' for char in characters]))
        
```

#### Generate and visualize valid augmentations by running the following code block:

```python
from rdkit import Chem

atom_drop_ratio = 0.2

aug = augmentSmiles(example_molecule, atom_drop_ratio)

visualize_smiles(aug)
```
