```python
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
from rdkit.Chem.EnumerateStereoisomers import EnumerateStereoisomers, StereoEnumerationOptions
```

```python
ibuprofen = AllChem.MolFromSmiles('CC(C)Cc1ccc(cc1)[C@@H](C)C(=O)O')
```

```python
opts = StereoEnumerationOptions(tryEmbedding=True, onlyUnassigned=False)

result = EnumerateStereoisomers(ibuprofen, options=opts)
```

```python
mols = []
for s_mol in result:
    AllChem.Compute2DCoords(s_mol)
    mols.append(s_mol)
```

```python
Draw.MolsToGridImage(mols, subImgSize=(500,500), useSVG=False)
```
